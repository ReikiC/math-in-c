# 基础运算

用 C 验证基本算术运算。修改数值后重新运行单元格即可看到结果。

## 加减乘除

设 $a = 7$, $b = 3$，计算四则运算：

```c
#include <stdio.h>

int main() {
    int a = 7, b = 3;
    printf("a + b = %d\n", a + b);
    printf("a - b = %d\n", a - b);
    printf("a * b = %d\n", a * b);
    printf("a / b = %d\n", a / b);
    printf("a %% b = %d\n", a % b);
    return 0;
}
```

## 浮点精度

比较 $\frac{1}{3}$ 的单精度与双精度表示：

```c
#include <stdio.h>

int main() {
    float f = 1.0f / 3.0f;
    double d = 1.0 / 3.0;
    printf("float:  %.9f\n", f);
    printf("double: %.15f\n", d);
    return 0;
}
```

## 勾股定理

验证 $3^2 + 4^2 = 5^2$，并计算斜边 $c = \sqrt{a^2 + b^2}$：

```c
#include <stdio.h>
#include <math.h>

int main() {
    double a = 3.0, b = 4.0;
    double c = sqrt(a*a + b*b);
    printf("c = %.6f\n", c);
    return 0;
}
```
